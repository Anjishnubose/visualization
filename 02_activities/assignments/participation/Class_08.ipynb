# Activity

This diagram of the British slave ship Brooks is one of the most powerful examples of 18th-century data visualization used for advocacy. It leverages rational, moral, and emotional appeals to influence public opinion against the transatlantic slave trade.

## Rational Appeal (Logos):

- The diagram provides a detailed, schematic representation of how enslaved Africans were packed into the ship's hold.
- It uses measurements, cross-sections, and numerical data (e.g., “454 slaves” across different decks) to highlight the inhumane conditions through an objective lens.
- By applying reason and empirical evidence, it undermines the claim that slaves were treated well during transport, a common pro-slavery argument at the time.

## Moral Appeal (Ethos):

- The image implicitly challenges the viewer’s sense of ethics by showing the systematic dehumanization of Black bodies, represented not as individuals but as densely packed, anonymous figures.
- It raises moral questions: Can such treatment of human beings ever be justified? Does legal regulation (as referenced in the title) excuse ethical wrongdoing?
- It appeals to the viewer’s conscience and positions the abolitionist movement as morally authoritative.

## Emotional Appeal (Pathos):

- Visually, the diagram evokes horror and empathy. The sheer number of human figures packed into tight, coffin-like spaces makes the suffering almost tangible.
- Even without showing violence or blood, the visualization conveys oppression and dehumanization in a way that words often cannot.
- For contemporary audiences, this stark image served as a gut-punch—eliciting outrage, sorrow, and a call to action
