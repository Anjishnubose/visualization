# Accessible Data Visualization – Notes

## Why Accessibility Matters
- *Inaccessible visualizations exclude* people with disabilities.
- Accessibility = better communication and broader audience reach.

## Colour
- Use **luminance**, **saturation**, and **hue** appropriately.
- Use **high contrast** for readability.
- Don’t rely on colour alone. Add *patterns or textures*.
- Tools:
  - [Coblis Color Blindness Simulator](https://www.color-blindness.com/coblis-color-blindness-simulator/)
  - [WebAIM Contrast Checker](https://webaim.org/resources/contrastchecker/)
  - *Viridis* palette: optimized for colourblind accessibility.

## Text & Labels
- Use *clear, consistent labels* and *legends*.
- *Sans-serif fonts* (e.g., Arial, Verdana) preferred for readability.
- *Min font size:* 12pt; use *monospaced text* where possible.
  
## Alt Text & Descriptions
- *Alt-text* is essential for screen reader users.
- Should include:
  - What the chart shows (title, axes)
  - Patterns or trends
  - Summary stats
  - Context if needed
- Avoid too much interpretation—let users draw conclusions.

## Accessibility Beyond Design
- Ensure *cross-browser compatibility*.
- Reduce *cost/access barriers* (use Open Access, CC licenses).
- Get *user feedback* to improve.
