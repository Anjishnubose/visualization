```python
##### Importing libraries #####
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
##### Random data points generation #####
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100,50)
```

```python
##### basic scatterplot #####
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y)
```

```python
##### Bar plot of the same data #####
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x,y)
```

```python
##### Line plot of the same data #####
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
```

```python
##### Histogram of the same data #####
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y, bins=10)
```

```python
##### Adding plot labels #####
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
fig
```

```python
##### Label properties #####
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

```python
##### Setting label locations #####
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

```python
##### Adding color and marker arguments to scatterplot #####
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter( x,
y,
marker='*', # NEW ADDITIONS
color = "indigo") # NEW ADDITIONS
fig.show()
```

```python
##### Adjusting line properties #####
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,
y,
marker='*',
color = 'indigo',
linestyle = '--', # NEW ADDITIONS
linewidth = 2) # NEW ADDITIONS
fig.show()
```

```python
##### Adjusting color using hexcodes #####
fig, ax = plt.subplots(figsize=(5, 3))
# color being the new addition
ax.plot(x,y,marker='*', color = '#7425b9', linestyle = '--', linewidth = 2)
fig.show()
```

```python
##### more keyword arguments #####
ax.plot(x,y,marker='*',
markersize = 12, # NEW ADDITIONS
color = '#7425b9',
linestyle = '--',
linewidth = 2,
markeredgecolor = '#fa9359', # NEW ADDITIONS
markerfacecolor = '#fa9359' ) # NEW ADDITIONS
fig
```

```python
##### Adding gridlines which are blue and dashed #####
ax.grid(axis = 'y', color = "blue", linewidth = 2, linestyle = '-.')
fig
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##### Importing data #####
df = sns.load_dataset('iris')
```

```python
sns.set_theme(style="darkgrid")
sns.kdeplot(df['sepal_width'], fill=True, color="r")
sns.kdeplot(df['sepal_length'], fill=True, color="b")
plt.show()
```

- The plot clearly distinguishes between the two datasets by plotting them in different colors, and hence distinguishing the two data trends.
