# Activity-1

The two visualizations differ in both purpose and design.

## The bar chart (Left)
- The bar chart shows average student ratings based on actual and perceived instructor gender. It focuses on *numerical comparison* and highlights potential bias in teaching evaluations.
- **Pros**: Clear, easy to compare values; good for precise interpretation.
- **Cons**: Limited scope; lacks context or qualitative insight.

## The dot plot (Right)
- The dot plot shows gendered language in reviews across disciplines. It emphasizes *language patterns* and *stereotypes*, using color to distinguish male/female word use.
- **Pros**: Rich insight into language use; interactive and detailed.
- **Cons**: Can be overwhelming; less intuitive without context.


# Activity-2

In the dynamic visualization
- The x-axis scale of occurences changes, the y-axis order of subjects changes, and of course the actual data being plotted changes based on the input terms.
- The title, axis labels, and data labels stay the same always.
- I would just want to speed up the animation when new terms are input into the visualization.

```python
import plotly.graph_objects as go # 'go' is 'graph objects’
x1 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
```

```python
graph = go.Figure()
graph.add_trace(go.Bar(x=x1, y=y1))
graph.update_layout(
title="Pirate Scores",
xaxis_title="Pirates",
yaxis_title="Score")
graph.show()
```

```python
graph.write_html("./pirategraph.html")
```

```python
graph = go.Figure()
graph.add_trace(go.Scatter(x=x1, y=y1, mode='markers', # we want points for a scatter plot
marker=dict(
size=15, # point size
color='hotpink', # point colour
opacity=1, # point transparency/alpha
line=dict(width=5, color='purple') # point outline
)))
graph.update_layout(
title='Interactive Pirate Plot',
xaxis_title='Pirates',
yaxis_title='Scores',
width=500, height=500)
```

```python
import wordcloud
from wordcloud import WordCloud
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv",
on_bad_lines='skip')
df
```

```python
# join all our text from each row from our quote column into a string
text = " ".join(each for each in df.quote)
# generate our wordcloud image
wordcloud = WordCloud(background_color="white",
colormap = 'inferno').generate(text)
# use matplotlib syntax to put our image in a figure
fig, ax = plt.subplots(figsize=(7, 3))
ax.imshow(wordcloud, # remember 'imshow' from when we added pictures to our matplotlib axes
interpolation='bilinear') # this line helps smooth our image
ax.axis("off")
plt.show()
```

```python
import matplotlib_venn
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
A = set(["apple", "banana", "watermelon"])
B = set(["pumpkin", "blueberry", "apple", "key lime"])
```

```python
diagram = venn2_unweighted([A, B],
set_labels = ('Fruits', 'Pies'),
set_colors=("blue", "red"),
alpha=0.5)
plt.show()
```

```python
diagram = venn2_unweighted([A, B],
set_labels = ('Fruits', 'Pies'),
set_colors=("blue", "red"),
alpha=0.5)
diagram.get_label_by_id("10").set_text("n".join(A - B))
diagram.get_label_by_id("11").set_text("n".join(A & B))
diagram.get_label_by_id("01").set_text("n".join(B - A))
plt.show()
```
